### How important is color really?
1. Color can reveal hidden elements.
3. Food ripening is also determined by color

### What about color blind folks?
1. How do they see viz?
2. How do they drive?
   Traffic signs are color based but they also have shapes and position of colors. Ex: road signs and traffic signals order
3. How do they leanr colors?
   Associate something with colors. Emotional sentiments.Red, white-soft/clouds, Blue-water. Etc.Associate color names to objects, touch or sentiments.

### Color recognition
1. Can't use short wavelength color since it's closer to dark colors and difficult to distinguish after dark.
2. Blue is not used since it's harder to identify than the color red. (with a Black background)

### How to choose the right color
1. Diverging colors for dividing. 
2. Choice of colors can hide information. 
3. Colors can be chosen for sequential and categorical
   For continous values, it might be best to use sequential colors.
   For discrete, use categorical.

### Example of when to choose seq
1. Viz of how many time do you kiss a friend when you meet by country?
2. This is largely sequential data and can be better respresneted by increasing contrast of a single color. 
3. Category is (1,2,3,4,5)

### Choose colors and make them pop
1. Several good color wheel tools online
2. Edges can be a good representative. *Jack Sparrow*
3. Luminance, contrast can be modified to make an image pop

### DOs and DONTs
1. Blue against black is a huge No. Difficult to read. 10:1 ration for text.
2. Larger the area to cover, smaller the saturation. Ex: Small lines/blocks will have high saturation colors.
3. Highlight data subsets (Ex: regions within a map)
4. Use consisent and standardiszed color palettes
5. In case of visual search, make sure you're not looking for a conjuction of features (both color and shape).

### Problems with categorical data
1. Categorical color schemes do not usually represent data very well. One example could be categorising number of times you kiss a friend when you greet them by country.
2. While bucketing is one way to do it, Sequential can be used in this case to highlight this.
3. Color is only for differences not for actual representation.  Objects are used for this purpose.

### Interactive plots

1. Interactive plots help you slice and dice with data filters and metrics.
2. Large amount of data becomes increasingly hard to visualize even if users are provided with filters and interactive plots.
3. We need some mechanism to process the data for selection, feedback and transparency of data exploration. 
4. Ex: Google motion charts, Riot rumours, Moneyball

### Difficult to browse and locate stuff

#### Enable selection
Google maps is a good example. The key here is search. You search with actual names (not raw co-ordinates). Or semantic queries like "where can I find ramen?". They don't just pin the places on the map viz. they list the places on the left and the surrounding areas to give you an idea of whether you want to explore that area.

Another example is the top view of a concert area when booking tickets. Users can take a call between lowest price ticket and the adjacent areas (like how close you are to the actual stage)

### Plotting

```r
library(ggplot2)
library(gdata)
```

```r
sample_file = read.csv("sample.csv")
```

```r
p = ggplot(sample_file, aes(winner, error))
p + geom_point()
```

```r
p = ggplot(sample_file, aes(winner, error, color=factor(victory), size=total, group=victory))
p + geom_point()
```

```r
myPalette = c("#FF9900", "#0099FF")
p + geom_point() + scale_colour_manual(values=myPalette)
```

```r
p = ggplot(sample_file, aes(winner, error, colour=factor(victory), size=total, label=player)) 
p + geom_point() +scale_colour_manual(values=myPalette) + geom_text()
```

```r
p + geom_point() +scale_colour_manual(values=myPalette) + geom_text(hjust=-0.2, vjust=1)
```

```r
p = ggplot(sample_file, aes(winner, error))
p + geom_point(aes(shape=factor(victory), size=total))+
geom_text(aes(colour=factor(year),label=player), hjust=1.2, vjust=-1)
```

```r
p + geom_point(aes(shape=factor(victory), size=total))+ 
geom_text(aes(colour=factor(year),label=player), position = position_jitter(width=5, height=1.5) ) + facet_grid(~year)
```
